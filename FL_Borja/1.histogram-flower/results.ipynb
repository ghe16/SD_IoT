```python
import pickle as pk
import numpy as np

with open("./histogram.pk", "rb") as f:
    data = pk.load(f)

bins = np.linspace(2.0, 10.0, 10)
bins = [np.mean([bins[i], bins[i+1]]) for i in range(len(bins)-1)]

data
```

```python
import matplotlib.pyplot as plt

plt.bar(bins, data["SepalLengthCm"], alpha=0.5, color="blue", edgecolor="black")
```

```python
import matplotlib.pyplot as plt

plt.bar(bins, data["SepalWidthCm"], alpha=0.5, color="blue", edgecolor="black")
```
